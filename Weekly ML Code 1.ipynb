```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('cell_phones.csv')
df2 = pd.read_csv('population1.csv')
df3 = pd.read_csv('ddf--entities--geo--country.csv')
df2.columns
```

```python
new_df1 = df1.merge(df2)
new_df1.shape
```

```python
df3.columns
```

```python
df3 = df3.rename(columns={"country":"geo", "name":"Country"})
print(df3.columns)
```

```python
df3_newcolumns = df3[['geo', 'Country']]
new_df1 = new_df1.merge(df3_newcolumns)
print(new_df1.columns)
new_df1.shape
new_df1.head()
```

```python
total_cell = new_df1['cell_phones_total']
total_population = new_df1['population']
cell_per_person = total_cell/total_population
new_df1['cell_per_person'] = cell_per_person
new_df1.head()
```

```python
new_df1['geo'] = new_df1['geo'].str.upper()
new_df1.head()
```
